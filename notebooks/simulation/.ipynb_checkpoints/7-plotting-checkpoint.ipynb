```python
def plot_roc_auc(key, model_roc_auc):
    scale, length = key.split('_')
    _, ax = plt.subplots()
    plt.suptitle(f"{scale} scale, seq length {length}")
    ax.set(title="ROC-AUC", xlabel=r"$\lambda$", xscale="log")

    model = "true"
    roc_auc_arr = np.load(f"{path_scores}/roc_auc_{model}_{key}.npy")
    roc_auc_mean = np.nanmean(roc_auc_arr)
    ax.axhline(roc_auc_mean, label="true")

    for model in model_roc_auc:
        roc_auc_arr = np.load(f"{path_scores}/roc_auc_{model}_{feat_name}_{key}.npy")
        roc_auc_mean = np.nanmean(roc_auc_arr, axis=1)
        ax.plot(lambd_list, roc_auc_mean, label=model)

    ax.legend()
    plt.savefig(f"{path_figures}/roc_auc_{key}.jpeg", bbox_inches="tight")
    return ax
```

```python
def plot_acc(key, model_accuracy):
    scale, length = key.split('_')
    _, ax = plt.subplots(*(1, 3), figsize=(40, 10))
    plt.suptitle(f"{scale} scale, seq length {length}")
    ax[0].set(title="Accuracy 0", xlabel=r"$\lambda$", xscale="log")
    ax[1].set(title="Accuracy 1", xlabel=r"$\lambda$", xscale="log")
    ax[2].set(title="Balanced accuracy", xlabel=r"$\lambda$", xscale="log")
    # accuracy_dict = {}
    # roc_auc_dict = {}
    model = "true"
    accuracy_arr = np.load(f"{path_scores}/accuracy_{model}_{key}.npy")
    acc_mean = np.nanmean(accuracy_arr, axis=0)
    ax[0].axhline(acc_mean[0], label="true")
    ax[1].axhline(acc_mean[1], label="true")
    ax[2].axhline(acc_mean.mean(), label="true")

    for model in model_accuracy:
        accuracy_arr = np.load(f"{path_scores}/accuracy_{model}_{feat_name}_{key}.npy")
        acc_mean = np.nanmean(accuracy_arr, axis=1)
        ax[0].plot(lambd_list, acc_mean[:, 0], label=model)
        ax[1].plot(lambd_list, acc_mean[:, 1], label=model)
        ax[2].plot(lambd_list, acc_mean.mean(axis=1), label=model)

    for ax_ in ax:
        ax_.legend()
    plt.savefig(f"{path_figures}/accuracy_{key}.jpeg", bbox_inches="tight")
    return ax
```

```python
model_accuracy = ["cont_mode", "cont_no_mode", "discrete"]
model_roc_auc = ["cont_mode", "cont_no_mode"]
```

```python
for key in key_list[:3]:
    plot_acc(key, model_accuracy)
    plot_roc_auc(key, model_roc_auc)
```

```python
model_accuracy = ["cont_mode", "discrete"]
model_roc_auc = ["cont_mode"]
for key in key_list[3:]:
    plot_acc(key, model_accuracy)
    plot_roc_auc(key, model_roc_auc)
```
